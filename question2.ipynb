```python
import pandas as pd
import datetime

N = 2

def dataread():
    log_records = pd.read_csv("log.csv",header=None)
    log_records = log_records.rename(columns={0:"Record"})
    return log_records

def main():
    log_records = dataread()
    trouble_start_dtes = {}
    trouble_count = {}
    
    for record in log_records["Record"]:
        first_comma_position = record.find(",")
        second_comma_position = record.rfind(",")
        str_dte = record[:first_comma_position]
        dte = datetime.datetime.strptime(str_dte, "%Y%m%d%H%M%S")
        server_address = record[first_comma_position+1:second_comma_position]
        response_result = record[second_comma_position+1:]
        
        if response_result == "-":
            if server_address not in trouble_count:
                trouble_count[server_address] = 1   
            else:
                trouble_count[server_address] += 1
                if trouble_count[server_address] == N:
                    trouble_start_dtes[server_address] = dte
        else:
            if server_address in trouble_start_dtes:
                trouble_start_dte = trouble_start_dtes.pop(server_address)
                recovery_dte = dte
                trouble_period = recovery_dte - trouble_start_dte
                print(server_address + "が復旧　故障期間:" + str(trouble_period))
            if server_address in trouble_count:
                del trouble_count[server_address]
            
                
        if trouble_start_dtes:
            trouble_server_addresses = []
            for trouble_server_address in trouble_start_dtes.keys():
                trouble_server_addresses.append(trouble_server_address)
            print(*trouble_server_addresses)
        else:
            print("故障無し")
            
if __name__=='__main__':    
    main()
```
